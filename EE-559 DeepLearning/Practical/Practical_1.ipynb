```
#Exercise 2 Eigendecomposition
M = torch.empty(20,20).normal_()
M_diag = torch.diag(M)
B=torch.inverse(M)*M_diag*M
B_eig = torch.eig(B,False)
```

```
#The returns of torch.eig is a namedtuple (eigenvalues, eigentvectos)
#eigenvalues(tensor): shape(n*2). Each row is an eigenvalue of input, where the
#first element is the real part and second element is the imaginary part
print(B_eig)
```

```
#Exercise 3 Flops per second
A = torch.empty(5000,5000).normal_()
B = torch.empty(5000,5000).normal_()
start = time.perf_counter()
torch.mm(A,B)
end = time.perf_counter()
num = 5000*5000*5000/(end-start)
print(num)
```

```
#Exercise 4 Playing with strides
def mul_row(m):
  result = torch.empty(m.size())
  for i in torch.arange(m.size()[0]):
    for j in torch.arange(m.size()[1]):
      result[i,j] = (i+1)*m[i,j]
  return result

def mul_row_fast(m):
  t = torch.arange(m.size()[1])+1
  result = torch.mul(m,t)
  return result

#torch.mul(input,other,out=None)
#multiplies each element of the input with the scalar other and returns a new
#resulting tensor
```

```
m = torch.empty((1000,400)).normal_()
```

```
start_1 = time.perf_counter()
mul_row(m)
end_1 = time.perf_counter()

start_2 = time.perf_counter()
mul_row_fast(m)
end_2 = time.perf_counter()
```

```
(end_1-start_1)/(end_2-start_2)
```
